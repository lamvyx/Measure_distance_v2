```python
import cv2
import os

# ThÆ° má»¥c chá»©a áº£nh gá»‘c
input_folder = "data/train/anhbandau"
# ThÆ° má»¥c lÆ°u áº£nh sau khi resize
output_folder = "data/train/images1"

# Táº¡o thÆ° má»¥c output náº¿u chÆ°a cÃ³
os.makedirs(output_folder, exist_ok=True)

# Duyá»‡t qua toÃ n bá»™ file trong thÆ° má»¥c
for filename in os.listdir(input_folder):
    # ÄÆ°á»ng dáº«n Ä‘áº§y Ä‘á»§ tá»›i file áº£nh
    input_path = os.path.join(input_folder, filename)

    # Kiá»ƒm tra pháº§n má»Ÿ rá»™ng há»£p lá»‡ (jpg, png, jpeg, v.v.)
    if filename.lower().endswith((".jpg", ".jpeg", ".png", ".bmp")):
        # Äá»c áº£nh
        img = cv2.imread(input_path)

        if img is None:
            print(f"âš ï¸ KhÃ´ng Ä‘á»c Ä‘Æ°á»£c áº£nh: {filename}")
            continue

        # Resize vá» 640x640
        resized = cv2.resize(img, (640, 640))

        # ÄÆ°á»ng dáº«n lÆ°u áº£nh má»›i
        output_path = os.path.join(output_folder, filename)

        # LÆ°u áº£nh Ä‘Ã£ resize
        cv2.imwrite(output_path, resized)
        print(f"âœ… ÄÃ£ xá»­ lÃ½: {filename}")

print("ğŸ¯ HoÃ n táº¥t resize táº¥t cáº£ áº£nh trong thÆ° má»¥c!")
```
